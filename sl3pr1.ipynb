```python
print(df.dtypes)
# Convert the target variable (species) to categorical variable
df[4] = df.iloc[:, -1].astype('category')
# Check the data types after conversion
print(df.dtypes)
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x = df.iloc[ : , :4]
x_scale = min_max_scaler.fit_transform(x)
df_norm = pd.DataFrame(x_scale)
df_norm
```

```python
df['Species'].unique()
```

```python
data_label = preprocessing.LabelEncoder()
df['Species'] = data_label.fit_transform(df['Species'])
df['Species'].unique()
```

```python
df.head()
```

```python
# One-hot encode the target variable (species)
df = pd.get_dummies(df, columns=[4])
# Display the first few rows of the dataframe after encoding
print(df.head())
```
